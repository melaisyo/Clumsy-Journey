### Numerical Elements -- The Duffing Oscillator (70pts)

Consider the following equation of motion for a damped and driven oscillator that is not necessarily linear. This is the [Duffing equation](https://en.wikipedia.org/wiki/Duffing_equation), which describes a damped driven simple harmonic oscillator with a cubic non-linearity. It's equation of motion is given by:

$$\ddot{x} + \delta \dot{x} + \alpha x + \beta x^3 = \gamma \cos(\omega t).$$

where $\alpha$, $\beta$, $\delta$, $\gamma$, and $\omega$ are constants.

Below is a figure showing the [strange attractor](https://en.wikipedia.org/wiki/Attractor#Strange_attractor) of the Duffing Oscillator over four periods. **It is locally chaotic, but globally it is stable.** This is a common feature of non-linear systems.



We will focus on the one dimensional case and we will analyze it in parts to put together a full picture of the dynamics of the system. Recall that the potential for a system can give rise to different kinds of behavior depending on the parameters (e.g. $\alpha$ and $\beta$).

b. (3pts) Find the equilibrium points of the system for the undriven and undamped case (assuming $m=1kg$). What are the conditions for stable and unstable equilibrium points? Plot the potential for some choices of $\alpha$ and $\beta$ to illustrate your findings of the conditions of stability. You can use (i) $\alpha=1$ and $\beta=1$, (ii) $\alpha=1$ and $\beta=-1$, (iii) $\alpha=-1$ and $\beta=1$, and (iv) $\alpha=-1$ and $\beta=-1$. *Hint: Can $x$ be an imaginary value?*

```python
#Part_B_3pts


# Define the potential function
def potential(x, alpha, beta):
    return (alpha / 2) * x**2 + (beta / 4) * x**4

# Define the equilibrium conditions
def equilibrium_points(alpha, beta):
    # Solve for dV/dx = 0
    return np.roots([beta, 0, alpha, 0])  # beta*x^3 + alpha*x = 0

# Generate data and plot potentials
x_vals = np.linspace(-2, 2, 500)
parameters = [
    (1, 1),   # (alpha, beta)
    (1, -1),
    (-1, 1),
    (-1, -1)
]

# Plotting
plt.figure(figsize=(12, 8))
for alpha, beta in parameters:
    V_vals = potential(x_vals, alpha, beta)
    equilibria = equilibrium_points(alpha, beta)
    stable_eq = []
    unstable_eq = []
    for eq in equilibria:
        if np.isreal(eq):
            eq = np.real(eq)
            second_derivative = alpha + 3 * beta * eq**2
            if second_derivative > 0:
                stable_eq.append(eq)
            else:
                unstable_eq.append(eq)
    plt.plot(x_vals, V_vals, label=f"$\\alpha={alpha}, \\beta={beta}$")
    for eq in stable_eq:
        plt.scatter(eq, potential(eq, alpha, beta), color="green", label="Stable Equilibrium", zorder=5)
    for eq in unstable_eq:
        plt.scatter(eq, potential(eq, alpha, beta), color="red", label="Unstable Equilibrium", zorder=5)

# Add labels and legend
plt.title("Potential $V(x)$ for Different Values of $\\alpha$ and $\\beta$")
plt.xlabel("$x$")
plt.ylabel("$V(x)$")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```
